```python
import csv
import pandas as pd 
from IPython.core.display import HTML

training_dir = "../../datasets/OCDC/training"
csv_file_path = "{}/ocdc_training_accuracy_loss_all.csv".format(training_dir)
df = pd.read_csv(csv_file_path)

transformation = 'standard_9_operations'
train = df.loc[(df['phase'] == 'train') & (df['augmentation'] == transformation)]
test = df.loc[(df['phase'] == 'test') & (df['augmentation'] == transformation)]

print("'{}' mean accuracy: {}".format(transformation, test['accuracy'].mean()))
print("'{}' standard deviation: {}".format(transformation, test['accuracy'].std()))
print("'{}' best loss: {}".format(transformation, test['loss'].min()))

train_best_accuracy = train[train['accuracy']==train['accuracy'].max()]
display(HTML(train_best_accuracy.to_html()))

test_best_accuracy = test[test['accuracy']==test['accuracy'].max()]
display(HTML(test_best_accuracy.to_html()))

test_best_loss = test[test['loss']==test['loss'].min()]
display(HTML(test_best_loss.to_html()))

# Performance Learning Curves: Learning curves calculated on the metric by which the model will be evaluated and selected, e.g. accuracy.
# Optimization Learning Curves: Learning curves calculated on the metric by which the parameters of the model are being optimized, e.g. loss.
df_accuracy_lines = pd.DataFrame({
    'accuracy train {:.3f} ({:.3f} ± {:.3f})'.format(train['accuracy'].max(), train['accuracy'].mean(), train['accuracy'].std()): train['accuracy'].to_numpy(),
    'accuracy test {:.3f} ({:.3f} ± {:.3f})'.format(test['accuracy'].max(), test['accuracy'].mean(), test['accuracy'].std()): test['accuracy'].to_numpy(),
    'loss train {:.3f} ({:.3f} ± {:.3f})'.format(train['loss'].min(), train['loss'].mean(), train['loss'].std()): train['loss'].to_numpy(),
    'loss test {:.3f} ({:.3f} ± {:.3f})'.format(test['loss'].min(), test['loss'].mean(), test['loss'].std()): test['loss'].to_numpy()
}, index = train['epoch'])

txt_title = "Standard: all transformations (9) ({:.3f} ± {:.3f} / {:.3f} ± {:.3f})".format(test['accuracy'].mean(), test['accuracy'].std(), test['loss'].mean(), test['loss'].std())
accuracy_lines = df_accuracy_lines.plot.line(ylim=(0,1), figsize=(20,10), grid=True, xticks=range(0,401,10), title=txt_title)
accuracy_lines.plot([test_best_accuracy['epoch'].values[0]], [test_best_accuracy['accuracy'].values[0]], 'go', c='r')
accuracy_lines.plot([test_best_loss['epoch'].values[0]], [test_best_loss['loss'].values[0]], 'go', c='r')
accuracy_lines.axhline(test['accuracy'].mean(), color="red", linestyle="--")
accuracy_lines.axhline(test['loss'].mean(), color="red", linestyle="--")
#accuracy_lines.get_figure().savefig("{}/plot_200.png".format(training_dir))

#display(HTML(test_no_augmentation.to_html()))




x = pd.DataFrame(columns=('epoch', 'acc_train', 'acc_test', 'loss_train', 'loss_test'))
x['epoch'] = train['epoch'].to_numpy()
x['acc_train'] = train['accuracy'].to_numpy()
x['loss_train'] = train['loss'].to_numpy()
x['acc_test'] = test['accuracy'].to_numpy()
x['loss_test'] = test['loss'].to_numpy()
#print(x)

#print("Best accuracy: {}".format(test['accuracy'].max()))
#print("Mean accuracy: {}".format(test['accuracy'].mean()))
#print("Standard deviation: {}".format(test['accuracy'].std()))
#print("Best loss: {}".format(test['loss'].min()))
#print(test['accuracy'].size)

latex_csv_file_path = csv_file_path.replace(".csv", "__{}_latex.csv".format(transformation))
x.to_csv(latex_csv_file_path, encoding='utf-8', index=False)
```
