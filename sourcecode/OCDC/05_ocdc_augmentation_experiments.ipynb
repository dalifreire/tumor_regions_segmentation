```python
from ocdc_dataloader import *
import torchvision.utils as vutils
from torchvision import transforms
from torchvision import utils

dataset_dir = "../../datasets/OCDC"

batch_size = 2
patch_size = (640, 640)
color_model = "LAB"
dataloaders = create_dataloader(tile_size="{}x{}".format(patch_size[0], patch_size[1]),
                                batch_size=batch_size, 
                                shuffle=True,
                                img_input_size=patch_size,
                                img_output_size=patch_size,
                                dataset_dir=dataset_dir,
                                color_model=color_model,
                                augmentation_strategy="random",
                                start_epoch=1,
                                validation_split=0.0)

dataset_train_size = len(dataloaders['train'].dataset)
dataset_test_size = len(dataloaders['test'].dataset)



images, masks, fname, original_size = next(iter(dataloaders['train']))

original_img = images[0].squeeze(0).cpu()
target_img = images[1].squeeze(0).cpu()

original_img_rgb = lab_to_rgb(pil_to_np(transforms.ToPILImage()(original_img)))
target_img_rgb = lab_to_rgb(pil_to_np(transforms.ToPILImage()(target_img)))


threshold = 0.86
#original_img_psc_l = original_img[0][original_img[0] < threshold].numpy()
#original_img_psc_a = original_img[1][original_img[0] < threshold].numpy()
#original_img_psc_b = original_img[2][original_img[0] < threshold].numpy()

original_img_cbar_l = original_img[0][original_img[0] < threshold].numpy().mean()
original_img_cbar_a = original_img[1][original_img[0] < threshold].numpy().mean()
original_img_cbar_b = original_img[2][original_img[0] < threshold].numpy().mean()

#original_img_cbar_l = original_img[0][original_img[0] < threshold].numpy().mean()
#original_img_cbar_a = original_img[1][original_img[0] < threshold].numpy().mean()
#original_img_cbar_b = original_img[2][original_img[0] < threshold].numpy().mean()
target_img_cbar_l = target_img[0][target_img[0] < threshold].numpy().mean()
target_img_cbar_a = target_img[1][target_img[0] < threshold].numpy().mean()
target_img_cbar_b = target_img[2][target_img[0] < threshold].numpy().mean()

original_img_psc = np.copy(original_img_rgb)
original_img_psc[:, :, 0][(original_img[0] >= threshold)] = 0
original_img_psc[:, :, 1][(original_img[0] >= threshold)] = 0
original_img_psc[:, :, 2][(original_img[0] >= threshold)] = 0

target_img_psc = np.copy(target_img_rgb)
target_img_psc[:, :, 0][(target_img[0] >= threshold)] = 0
target_img_psc[:, :, 1][(target_img[0] >= threshold)] = 0
target_img_psc[:, :, 2][(target_img[0] >= threshold)] = 0

print("Original:")
print("{} -> {}".format(np.mean(original_img[0].numpy()), original_img_cbar_l)) # L*
print("{} -> {}".format(np.mean(original_img[1].numpy()), original_img_cbar_a)) # a*
print("{} -> {}".format(np.mean(original_img[2].numpy()), original_img_cbar_b)) # b*

print("target:")
print("{} -> {}".format(np.mean(target_img[0].numpy()), target_img_cbar_l)) # L*
print("{} -> {}".format(np.mean(target_img[1].numpy()), target_img_cbar_a)) # a*
print("{} -> {}".format(np.mean(target_img[2].numpy()), target_img_cbar_b)) # b*

augmented_img = np.copy(original_img.permute(1, 2, 0).numpy())
augmented_img[:, :, 0][(original_img[0] < threshold)] = augmented_img[:, :, 0][(original_img[0] < threshold)] - original_img_cbar_l + target_img_cbar_l
augmented_img[:, :, 1][(original_img[0] < threshold)] = augmented_img[:, :, 1][(original_img[0] < threshold)] - original_img_cbar_a + target_img_cbar_a
augmented_img[:, :, 2][(original_img[0] < threshold)] = augmented_img[:, :, 2][(original_img[0] < threshold)] - original_img_cbar_b + target_img_cbar_b
augmented_img = torch.from_numpy(augmented_img).permute(2, 0, 1)

#augmented_img = np.copy(original_img)
#augmented_img[0] = original_img[0].numpy() - original_img_cbar_l + target_img_cbar_l
#augmented_img[1] = original_img[1].numpy() - original_img_cbar_a + target_img_cbar_a
#augmented_img[2] = original_img[2].numpy() - original_img_cbar_b + target_img_cbar_b
#augmented_img = torch.from_numpy(augmented_img)

print("Augmented:")
print(np.mean(augmented_img[0].numpy()))
print(np.mean(augmented_img[1].numpy()))
print(np.mean(augmented_img[2].numpy()))
augmented_img_rgb = lab_to_rgb(pil_to_np(transforms.ToPILImage()(augmented_img)))

f, ax = plt.subplots(3, 3, figsize=(10, 10))
ax[0][0].imshow(original_img.permute(1, 2, 0))
ax[0][0].set_title('Original image L*a*b*')
ax[0][1].imshow(original_img_rgb)
ax[0][1].set_title('Original image RGB')
ax[0][2].imshow(original_img_psc)
ax[0][2].set_title('Original image RGB (PSC)')
#
ax[1][0].imshow(target_img.permute(1, 2, 0))
ax[1][0].set_title('Target image L*a*b*')
ax[1][1].imshow(target_img_rgb)
ax[1][1].set_title('Target image RGB')
ax[1][2].imshow(target_img_psc)
ax[1][2].set_title('Target image RGB (PSC)')
#
ax[2][0].imshow(original_img.permute(1, 2, 0))
ax[2][0].set_title('Augmented image L*a*b*')
ax[2][1].imshow(masks[0], cmap='gray')
ax[2][1].set_title('Mask image')
ax[2][2].imshow(augmented_img_rgb)
ax[2][2].set_title('Augmented image RGB')

np_to_pil(original_img_rgb).save('/home/dalifreire/Pictures/teste/{}_original.png'.format(fname[0]))
np_to_pil(original_img_psc).save('/home/dalifreire/Pictures/teste/{}_psc.png'.format(fname[0]))
np_to_pil(augmented_img_rgb).save('/home/dalifreire/Pictures/teste/{}_augmented.png'.format(fname[0]))
```
